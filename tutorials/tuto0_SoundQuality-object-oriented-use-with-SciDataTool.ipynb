# How to use SoundQuality with SciDatatool

MoSQITo is a unified and modular development framework of key sound quality metrics favoring reproducible science and efficient shared scripting among engineers, teachers and researchers community.

MoSQITo has been developed as a library of functions, but in order to ease its use, a class called SoundQuality has been created. Its main advantage is the use of [SciDataTool](https://github.com/Eomys/SciDataTool "SciDataTool") objects to store the results and do efficient plots.


This tutorial explains how to use the SoundQuality object to calculate the acoustic loudness (Zwicker method) of a time-varying signal from its .wav file.

```python
# Add MoSQITo to the Python path
import sys
sys.path.append('..')

# Import useful packages
import numpy as np
import matplotlib.pyplot as plt

# Import MoSQITo modules
from mosqito.SoundQuality import SoundQuality
```

First, it is necessary to create a SoundQuality object, which uses the MOSQITO functions to compute metrics that are then stored in SciDataTool objects.

```python
SQ = SoundQuality()
```

The signal is imported using the 'import_signal' method with 2 arguments, the signal type (True if stationary or False if time-varying) and the file path.

Then its third-octave spectrum is computed with the 'comp_3oct_spec' method.

```python
SQ.import_signal(False,"../mosqito/validations/loudness_zwicker/data/ISO_532-1/Annex B.5/Test signal 24 (woodpecker).wav")
SQ.comp_3oct_spec()
```

The loudness is computed using the 'compute_loudness' method.

```python
SQ.compute_loudness()
```

The calculation being done, the results can be plotted or used for further analysis. The SciDataTools objects have different interesting functions for that purpose.

```python
loudness = SQ.loudness.values
time = SQ.loudness.axes[0].values
plt.plot(time, loudness)
plt.show()
```
